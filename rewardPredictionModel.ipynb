```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

df = pd.read_csv("training_data.csv")
print(df)
```

```python
%matplotlib inline
plt.xlabel('timeStamp(sec)')
plt.ylabel('blockReward(ETH)')
plt.scatter(df.timestamp,df.blockReward,color="blue",marker=".")
```

```python
model = RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(df.timestamp.values.reshape(-1,1),df.blockReward.values.ravel())
```

```python
new_timestamp = np.array([[1675388399]])
predicted_blockReward = model.predict(new_timestamp)

plt.scatter(df.timestamp,df.blockReward,label="Data Points")
plt.scatter(new_timestamp,predicted_blockReward,color="red",label="Predicted Data")
plt.xlabel('TimeStamp')
plt.ylabel('BlockReward')
plt.legend()
plt.show()

print("Predicted Block Reward is :",predicted_blockReward[0])
```
